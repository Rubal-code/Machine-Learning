```python
import pandas as pd
import numpy as np

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

import seaborn as sns
```

```python
df=pd.read_csv('Titanic-Dataset.csv')[['Age','Parch','SibSp','Pclass','Survived']]
```

```python
df.head()
```

```python
df.isnull().mean()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df.isnull().mean()
```

```python
x=df.iloc[:,0:4]
y=df.iloc[:,-1]
```

```python
np.mean(cross_val_score(LogisticRegression(),x,y,scoring='accuracy',cv=20))
```

# apply feature construction

```python
x['family_size']=df['SibSp']+df['Parch']+1
```

```python
x.head()
```

```python
def myfunc(num):
    if num==1:
        #alone
        return 0
    elif num>1 and num<=4:
        # small family
        return 1
    else:
        # big family
        return 2
```

```python
myfunc(4)
```

```python
x['family_type']=x['family_size'].apply(myfunc)
```

```python
x
```

```python
x.drop(columns=['Parch','SibSp','family_size'],inplace=True)
```

```python
x.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),x,y,scoring='accuracy',cv=20))
```
