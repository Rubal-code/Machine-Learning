```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('customer_data.csv')
```

```python
df.head()
```

```python
df.iloc[:,2:]
```

```python
df.drop(columns=['income','purchase_amount','promotion_usage','satisfaction_score'],inplace=True)
```

```python
df.head()
```

```python
df.drop(columns=['id','age'],inplace=True)
```

```python
df.head()  
# gender -> nominal data so ethe one hot encoder use hoega but odee li transformer chahida hunda jeda hale tak mae nhi sikhya so drop that cols which are nominal
```

```python
df.drop(columns=['gender','region','product_category'],inplace=True)
```

```python
df.head()
# hun pehle do columns are ordinal data because  masters > Bachelor > college > highschool  |  gold > silver > regular  ->ordinalencoding

#purchase_freq is label means output so label encoder 
```

```python
df['education'].unique() ,df['loyalty_status'].unique(),df['purchase_frequency'].unique()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(df.drop(columns=['purchase_frequency']) , df['purchase_frequency'] , test_size=0.3)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.preprocessing import OrdinalEncoder

oe=OrdinalEncoder(categories=[['HighSchool' ,'College' ,'Bachelor' ,'Masters'] , ['Regular',"Silver",'Gold']])
# highschool->0  college->1   regular->0  apne app ainve karke values assign kardega

```

```python
oe.fit(x_train)
```

```python
x_train_transform = oe.transform(x_train)
x_test_transform = oe.transform(x_test)
```

```python
x_train_transform
```

```python
x_test_transform
```

```python
x_train_transform=pd.DataFrame(x_train_transform,columns=x_train.columns)
```

```python
x_train_transform.sample(10)
```

```python
# now for output column
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()      # ethe ni lod pass karn li kinu kedi value dvani aa eh ape kardega

le.fit(y_train)


y_train_transform = le.transform(y_train)
y_test_transform = le.transform(y_test)
```

```python
y_train_transform
```

```python
y_train_transform=pd.DataFrame(y_train_transform , columns=['loyalty_status'])
```
