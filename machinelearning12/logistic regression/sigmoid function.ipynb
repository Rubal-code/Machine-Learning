```python
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt
```

```python
X,y=make_classification(n_samples=100,n_features=2,n_informative=1,n_redundant=0,
                        n_classes=2,n_clusters_per_class=1,random_state=41,hypercube=False,class_sep=40)
```

```python
plt.figure(figsize=(14,7))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=80)
```

```python
def sigmoid(z):                    # step di jagah sigmoid
    return 1 / (1+np.exp(-z))
```

```python
def percepton(X,y):
    X=np.insert(X,0,1,axis=1)
    weights=np.ones(X.shape[1])
    lr=0.1

    for i in range(1000):
        j=np.random.randint(0,100)
        y_hat=sigmoid(np.dot(X[j],weights))
        weights=weights + lr * (y[j] - y_hat) * X[j]
    
    return weights[0],weights[1:]
```

```python
intercept,coef=percepton(X,y)
```

```python
intercept,coef
```

```python
m=-(coef[0] / coef[1])
b=-(intercept / coef[1])
```

```python
x_input=np.linspace(-3,3,100)
y_input=m*x_input+b
```

```python
plt.figure(figsize=(14,7))
plt.plot(x_input,y_input,c='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=80)
plt.ylim(-3,2)
plt.xlim(-8,8)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(X,y)
```

```python
m_lr=-(lr.coef_[0][0] / lr.coef_[0][1])
b_lr=-(intercept / lr.coef_[0][1])
```

```python
x_input_1=np.linspace(-3,3,100)
y_input_1=m_lr * x_input_1 + b_lr
```

```python
plt.figure(figsize=(14,7))
plt.plot(x_input,y_input,c='red',linewidth=3)
plt.plot(x_input_1,y_input_1,c='brown',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=80)
plt.ylim(-3,2)
plt.xlim(-8,8)
```
